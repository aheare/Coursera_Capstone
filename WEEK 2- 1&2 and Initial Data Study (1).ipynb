### Problem and Data Discription for a Means to Solve 

Millions of people are affected by vehicular accidents every year in the united states over six million! Of those accidents the severity of such accidents ranges from mild property damage, to disabling injuries and even death. In fact, the average death toll in the united states due to vehicular accidents is around ninety per day!

There are multiple reasons for traffic incidences with the vast majority being caused by human error. Distractions, speeding, inclement weather, and inhabilitated drivers. With this study I hope to provide insight as to which factors are the main contributors to these accidents and what strategies could be considered. 

This study is based on a data set provided by the Seattle Department of Transportation. It includes multiple variables and will be analyzed for specific correlations and relationships. That could potentially help solve or reduce the amount of accidents or possible eliminate accidents that cause serious injury and death.

```python
def bar_plot(data_column, x_label, y_label, title, size=(12,8), rotate=0):
    a = data_column.value_counts()
    codes = a.index
    count = a.values
    plt.figure(figsize=size)
    ax = sns.barplot(x = codes, y=count)
    ax.set_xlabel(x_label, fontsize=15)
    ax.set_ylabel(y_label, fontsize=15)
    ax.set_title(title, fontsize=22)
    plt.xticks(rotation=rotate)
    plt.show()
```

```python
df['INCDTTM'] = pd.to_datetime(df['INCDTTM'], infer_datetime_format=True)
df['Year'] = df['INCDTTM'].dt.year
df['Month'] = df['INCDTTM'].dt.month
df['Day'] = df['INCDTTM'].dt.weekday
```

```python

plt.figure(figsize=(12,8))
ax = sns.countplot(x='Year', hue='SEVERITYDESC', data=df,
                   hue_order=['Property Damage Only Collision', 'Injury Collision'], dodge=False)
ax.set_xlabel('Year', fontsize=15)
ax.set_ylabel('Number of Collisions', fontsize=15)
ax.set_title('Collision Severity by Year', fontsize=19)
ax.legend(title=' Weighted Severity ');
```

```python

months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.figure(figsize=(12,8))
ax = sns.countplot(x='Month', hue='SEVERITYDESC', data=df,
                   hue_order=['Property Damage Only Collision', 'Injury Collision'], dodge=False)
ax.set_xlabel('Month', fontsize=15)
ax.set_ylabel('Number of Collisions', fontsize=15)
ax.set_title('Collision Severity by Month', fontsize=22)
ax.set_xticklabels(months)
ax.legend(title='Severity Description', bbox_to_anchor=(1,1));
```
